## Performing linear regression with homemade gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Generating some data

```python
x = np.random.normal(0,3,700)
y = 2 + np.random.normal(x*4,3)
```

Plotting the data

```python
plt.subplot(2,1,1)
plt.hist(x, 20)

plt.subplot(2,1,2)
plt.scatter(x, y, alpha=0.2)
```

Starting with initial guesses for our coefficients and setting the form of the equation

```python
b1 = 3
b0 = 0
yHat = b1*x + b0
```

Defining error metric

```python
error = (y - yHat)**2   
SSE = error.sum().round(1)


print(SSE)
```

Plotting the data and our initial estimate. Close, but no cigar.

```python
sns.plt.plot(x, y, '.')
plt.plot(x, yHat,'r-');
```

```python
def gradient_step(current_b0, current_b1, x, y):
    b0_grad = 0 
    b1_grad = 0
    learn_rate = 0.05
    N = float(len(x))
    for i in range(0,len(x)):
        b0_grad += (-2/N)*(y[i] - ((current_b1*x[i]) + current_b0))
        b1_grad += (-2/N)*x[i]*(y[i] - ((current_b1*x[i]) + current_b0))
    new_b0 = current_b0 - learn_rate*b0_grad
    new_b1 = current_b1 - learn_rate*b1_grad
    return [new_b0, new_b1]

def simpLinReg(x,y):
    iterNum = 50
    b0 = 0
    b1 = 0 
    for i in range(iterNum):
        b0, b1 = gradient_step(b0,b1,x,y)
    return [b0, b1]
```
